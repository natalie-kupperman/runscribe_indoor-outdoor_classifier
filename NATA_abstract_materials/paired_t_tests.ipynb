```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd. read_csv('output/full_dataset.csv')
```

```python
df_maxpronation = df[['subject_id', 'medium', 'max_pronation_velocity']]
table_maxpronation = pd.pivot_table(df_maxpronation, values = 'max_pronation_velocity',
                                     columns = 'medium', index = 'subject_id',
                                     aggfunc = np.mean)
maxpro_t, maxpro_p = stats.ttest_rel(table_maxpronation['outdoor'], table_maxpronation['treadmill'])


df_pronationexcursionloading = df[['subject_id', 'medium', 'pronation_excursion_fs_mp']]
table_pronationexcursionloading = pd.pivot_table(df_pronationexcursionloading, values = 'pronation_excursion_fs_mp',
                                                 columns = 'medium', index = 'subject_id',
                                                 aggfunc = np.mean)
maxproloadex_t, maxproloadex_p = stats.ttest_rel(table_pronationexcursionloading['outdoor'], table_pronationexcursionloading['treadmill'])


df_supexcurprop = df[['subject_id', 'medium', 'pronation_excursion_mp_to']]
table_supexcurprop = pd.pivot_table(df_supexcurprop, values = 'pronation_excursion_mp_to',
                                     columns = 'medium', index = 'subject_id',
                                     aggfunc = np.mean)
supexcurprop_t, supexcurprop_p = stats.ttest_rel(table_supexcurprop['outdoor'], table_supexcurprop['treadmill'])


df_length = df[['subject_id', 'medium', 'step_length']]
table_length = pd.pivot_table(df_length, values = 'step_length',
                              columns = 'medium', index = 'subject_id',
                              aggfunc = np.mean)
length_t, length_p = stats.ttest_rel(table_length['outdoor'], table_length['treadmill'])


df_footstrike = df[['subject_id', 'medium', 'footstrike_type']]
table_footstrike = pd.pivot_table(df_footstrike, values = 'footstrike_type',
                                  columns = 'medium', index = 'subject_id',
                                  aggfunc = np.mean)

footstrike_t, footstrike_p = stats.ttest_rel(table_footstrike['outdoor'], table_footstrike['treadmill'])


df_velocity = df[['subject_id', 'medium', 'stride_pace']]
table_velocity = pd.pivot_table(df_velocity, values = 'stride_pace',
                                  columns = 'medium', index = 'subject_id',
                                  aggfunc = np.mean)

velocity_t, velocity_p = stats.ttest_rel(table_velocity['outdoor'], table_velocity['treadmill'])


df_cycletime = df[['subject_id', 'medium', 'cycle_time']]
table_cycletime = pd.pivot_table(df_cycletime, values = 'cycle_time',
                                  columns = 'medium', index = 'subject_id',
                                  aggfunc = np.mean)

cycletime_t, cycletime_p = stats.ttest_rel(table_cycletime['outdoor'], table_cycletime['treadmill'])


df_cadence = df[['subject_id', 'medium', 'step_rate']]
table_cadence = pd.pivot_table(df_cadence, values = 'step_rate',
                               columns = 'medium', index = 'subject_id',
                               aggfunc = np.mean)

cadence_t, cadence_p = stats.ttest_rel(table_cadence['outdoor'], table_cadence['treadmill'])


df_brakingg = df[['subject_id', 'medium', 'braking_gs']]
table_brakingg = pd.pivot_table(df_brakingg, values = 'braking_gs',
                                columns = 'medium', index = 'subject_id',
                                aggfunc = np.mean)

brakingg_t, brakingg_p = stats.ttest_rel(table_brakingg['outdoor'], table_brakingg['treadmill'])


df_contacttime = df[['subject_id', 'medium', 'contact_time']]
table_contacttime = pd.pivot_table(df_contacttime, values = 'contact_time',
                                   columns = 'medium', index = 'subject_id',
                                   aggfunc = np.mean)

contacttime_t, contacttime_p = stats.ttest_rel(table_contacttime['outdoor'], table_contacttime['treadmill'])


df_contactratio = df[['subject_id', 'medium', 'contact_ratio']]
table_contactratio = pd.pivot_table(df_contactratio, values = 'contact_ratio',
                                   columns = 'medium', index = 'subject_id',
                                   aggfunc = np.mean)

contactratio_t, contactratio_p = stats.ttest_rel(table_contactratio['outdoor'], table_contactratio['treadmill'])


df_flightratio = df[['subject_id', 'medium', 'flight_ratio']]
table_flightratio = pd.pivot_table(df_flightratio, values = 'flight_ratio',
                                   columns = 'medium', index = 'subject_id',
                                   aggfunc = np.mean)

flightratio_t, flightratio_p = stats.ttest_rel(table_flightratio['outdoor'], table_flightratio['treadmill'])


df_flighttime = df[['subject_id', 'medium', 'flight_time']]
table_flighttime = pd.pivot_table(df_flighttime, values = 'flight_time',
                                   columns = 'medium', index = 'subject_id',
                                   aggfunc = np.mean)

flighttime_t, flighttime_p = stats.ttest_rel(table_flighttime['outdoor'], table_flighttime['treadmill'])


df_impactg = df[['subject_id', 'medium', 'impact_gs']]
table_impactg = pd.pivot_table(df_impactg, values = 'impact_gs',
                               columns = 'medium', index = 'subject_id',
                               aggfunc = np.mean)

impactg_t, impactg_p = stats.ttest_rel(table_flighttime['outdoor'], table_flighttime['treadmill'])


print(f'Max Pronation Velocity: {maxpro_t}({maxpro_p})')
print(f'Pronation Excursion During Loading Response: {maxproloadex_t}({maxproloadex_p})')
print(f'Supination Excursion During Propulsion: {supexcurprop_t}({supexcurprop_p})')
print(f'Step Length: {length_t}({length_p})')
print(f'Footstrike Type: {footstrike_t}({footstrike_p})')
print(f'Velocity: {velocity_t}({velocity_p})')
print(f'Cycle Time: {cycletime_t}({cycletime_p})')
print(f'Cadence: {cadence_t}({cadence_p})')
print(f'Braking Gs: {brakingg_t}({brakingg_p})')
print(f'Contact Time: {contacttime_t}({contacttime_p})')
print(f'Contact Ratio: {contactratio_t}({contactratio_p})')
print(f'Flight Ratio: {flightratio_t}({flightratio_p})')
print(f'Flight Time: {flighttime_t}({flighttime_p})')
print(f'Impact gs: {impactg_t}({impactg_p})')
```
